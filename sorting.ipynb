```python
# Bubble Sort
# Time:O(n^2)
# Space:O(1) 


def bubble_sort(arr):
    n=len(arr)
    flag = True
    while flag:
        flag = False
        for i in range (1,n):
            if arr[i-1]>arr[i]:
                flag = True
                arr[i-1],arr[i] = arr[i],arr[i-1]
                print(arr)
    return arr
```

```python
#Insertion Sort
# Time:O(n^2)
# Space:O(1)


def insertion_sort(arr):
    n= len(arr)
    for i in range(1,n):
        for j in range(i,0,-1):
            if arr[j-1]>arr[j]:
                arr[j-1],arr[j] = arr[j],arr[j-1]
                print(arr)
            else:
                print("getting broken")
                break
            
    return arr
```
